```python
list_item = []
for column in dfdt.columns:
    list_item.append([column,
                    dfdt[column].dtype,
                    dfdt[column].isna().sum(),
                    (dfdt[column].isna().sum()/len(dfdt[column])*100),
                    dfdt[column].nunique(),
                    list(dfdt[column].unique()[:2])
                     ])
desc_df = pd.DataFrame(data = list_item, columns=['features', 'dtype', 'null', 'null%', 'unique', 'unique_sample'])
desc_df
```

```python
# function untuk mengimputasi rating berdasarkan rata-rata rating tiap Category
def impute_rating(cols):
    rating = cols[0]
    cate=cols[1]
    if pd.isnull(rating):
        return round(dfdt[(dfdt['Category'] == cate)].mean()[0],2)
    else:
        return rating
```

```python
dfdt['Rating'] = dfdt[['Rating', 'Category']].apply(impute_rating,axis=1)
```

```python
dfdt.dropna(inplace=True)
```

#### Size Feature

```python
idx=list(range(0,10829))
dfdt.set_index(pd.Index(idx), inplace=True)
dfdt
```

```python
dfdt.Size.unique()
```

Terlihat bahwa fitur Size bertipe string/object, disini maka akan dilakukan pengkonversian terhadap fitur Size, dimana data yang berakhiran k diubah menjadi 1000 dan M diubah menjadi 1000000, dan mengubah data Varies with device menjadi nan

```python
dfdt['Size']=dfdt.Size.str.replace('k','e+3')
dfdt['Size']=dfdt.Size.str.replace('M','e+6')
dfdt['Size']=dfdt.Size.replace('Varies with device',np.nan)
dfdt['Size'].head()
```

```python
dfdt['Size']=pd.to_numeric(dfdt['Size'])
```

```python
dfdt['Size']=dfdt['Size']/1000000
```

```python
dfdt['Size']=round(dfdt['Size'],2)
```

#### Install Feature

```python
dfdt['Installs'].value_counts()
```

Pada data Install ini nanti nya akan menjadi fitur yang bersifat ordinal dengan range banyaknya install

```python
dfdt.Installs=dfdt.Installs.apply(lambda x: x.strip('+'))
dfdt.Installs=dfdt.Installs.apply(lambda x: x.replace(',',''))
dfdt.Installs=dfdt.Installs.replace('Free',0)
dfdt.Installs.value_counts()
```

```python
dfdt.Installs=pd.to_numeric(dfdt.Installs)
```

#### Last Update Feature

Untuk fitur Last update akan diubah menjadi selisih dari tanggal update terakhir dengan tanggal data ini discrapping, sehingga bisa ketahui pengaruh dari update terhadap model nantinya

```python
from datetime import datetime,date
temp=pd.to_datetime(dfdt['Last Updated'])
temp
```

```python
scrapdate=date(2019, 5, 10)
scrapdate
```

```python
dfdt['Last Updated']=temp.apply(lambda x:scrapdate-datetime.date(x))
```

```python
dfdt['Last Updated']=dfdt['Last Updated'].astype(str).apply(lambda x: x.strip('days 00:00:00.000000000'))
```

```python
dfdt['Last Updated']=pd.to_numeric(dfdt['Last Updated'])
```

```python
dfdt.rename(columns = {'Last Updated':'Last Updated(days ago)'}, inplace = True)
```
